# Landscape classification, first idea

This notebook is my attempt to classify a database of sea, mountains, buildings, forests and roads pictures using the simpliest way to do a neural network on TensorFlow 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import skimage as sk
import os
import tensorflow as tf
print(tf.__version__)
tf.config.set_visible_devices([], 'GPU') # I am working on a Mac M1, this line is to crontol Tensorflow's access to the GPU
```

```python
# Loading the images, 
# they are named 1.jpg, 2.jpg, 3.jpg ... so my goal is to automatically get the number of every image so that
# we can add images in the file.

images = os.listdir("./input") 

images = [os.path.splitext(images_num)[0] for images_num in images]
images.sort()
images.pop(0)

images = np.array(images, dtype = int)
images.sort()

images
```

```python

data_set = np.zeros((len(images),256,256,3))
data_set.shape

# This loop is a little bit long but I did not find a better way to do it.

for i in range(len(images)) :
    data_set[i-1,:,:,:] = sk.io.imread("./input/"+str(images[i-1])+".jpg") / 255 
```

```python
np.info(data_set)
```

```python
# This portion of the code is here to label the data because the images were sorted by class but not labeled.

class_names = ["ocean & sea", "forest", "road", "building", "mountains"] # 0 : ocean & sea, 1 : forest, 2 : road ...

# I also separated the images into training and test sets. 
# First, I defined the size of every vector I am going to use.

label_set = np.zeros(len(images))

data_train = np.zeros((1352,256,256,3))
label_train = np.zeros(1352)

data_test = np.zeros((281,256,256,3))
label_test = np.zeros(281)

# Then, I matched every image with its label.

label_set[0:139] = 0
data_train[0:300,:,:,:] = data_set[:300,:,:,:]
label_train[0:300] = 0
label_train[140] = 1

label_set[140] = 1

label_set[141:358] = 0
data_test[0:58,:,:,:] = data_set[300:358,:,:,:]
label_test[0:58] = 0

label_set[359:686] = 1
data_train[300:541,:,:,:] = data_set[359:600,:,:,:]
data_test[58:143,:,:,:] = data_set[601:686,:,:,:]
label_train[301:542] = 1
label_test[58:143] = 1

label_set[687:946] = 2
data_train[541:754,:,:,:] = data_set[687:900,:,:,:]
data_test[143:188,:,:,:] = data_set[901:946,:,:,:]
label_train[542:755] = 2
label_test[143:188] = 2

label_set[947:1254] = 3
data_train[754:1007,:,:,:] = data_set[947:1200,:,:,:]
data_test[188:241,:,:,:] = data_set[1201:1254,:,:,:]
label_train[755:1008] = 3
label_test[188:241] = 3

label_set[1255:1642] = 4
data_train[1007:1352,:,:,:] = data_set[1255:1600,:,:,:]
data_test[241:281,:,:,:] = data_set[1601:1641,:,:,:]
label_train[1008:1352] = 4
label_test[241:281] = 4

# Then I check if there is no issue with the indices of the images in the data set and the train set. Such issues 
# would manifest as a black image, so I check the maximum pixel value to determine if its value is greater than zero.

for i in range(len(data_train)):
    if np.max(data_train[i,:,:,:]) == 0 :
       print(i)
    
for i in range(len(data_test)):
    if np.max(data_test[i,:,:,:]) == 0 :
       print(i)
    
label_test = label_test.astype(int)
label_train = label_train.astype(int)
```

```python
# Here is an insight of the images
plt.figure(figsize = (12,12))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(data_train[20*i,:,:,:])
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(class_names[label_train[20*i]])
```

```python
plt.figure(figsize = (12,12))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(data_test[11*i,:,:,:])
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(class_names[label_test[11*i]])
```

```python
# Here we build the model, Sequential is the first model I learned on TensorFlow so let's try with this one

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (256,256,3)), # This line convert the images into a column vector so that 
    # the model can process it. 
    tf.keras.layers.Dense(500, activation = "relu"), # I wanted a quite simple model to begin with so I chose to 
    tf.keras.layers.Dense(150, activation = "relu"), # put only 2 hidden layers
    tf.keras.layers.Dense(5)]) # Outpout layer of size 5, the number of class
# The Relu fonction is an activation function that does not change the value if it's positive, otherwise the value
# is O.

# Then we compile the model with the 'adam' optimizer, Adam seems to be the best one for this model as I tried 
# differents ones.
model.compile(optimizer = 'adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'])
```

```python
# Finally, we train the model
hist = model.fit(data_train, label_train, epochs=20,
                validation_data = (data_test, label_test))
```

```python
# Plotting important information to see how the model evolve and perform
pd.DataFrame(hist.history).plot(figsize=(8, 5)) 
plt.grid(True)
plt.gca().set_ylim(0, 1) 
plt.show()
```

# Conclusion 

We can clearly see that the model is not progressing well on the test set and seems to be overfitting the data as the difference between the train accuracy and test accuracy is more than 0.4 at some points.

The issue here is that even when I change the parameters of the model, it either becomes less accurate or overfits the data.

The next part is simply to observe how poorly the model performs in predicting the labels on the test set.

```python
test_loss, test_acc = model.evaluate(data_test, label_test, verbose = 2)
```

```python
prob_model = tf.keras.Sequential([
    model,
    tf.keras.layers.Softmax()])
```

```python
predictions = prob_model.predict(data_test)
print("True label : ", class_names[int(label_test[100])],"\n" )
print("Predicted label :",class_names[np.argmax(predictions[100])])
```

```python
def plot_image(i, predictions_array, true_label, img):
    true_label, img = true_label[i], img[i]
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)
    
    predicted_label = np.argmax(predictions_array)
    
    if predicted_label == true_label:
        color = 'green'
    else:
        color = 'red'
    
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)
def plot_value_array(i, predictions_array, true_label):
    true_label = true_label[i]
    plt.xticks(range(5))
    plt.yticks([])
    thisplot = plt.bar(range(5), predictions_array, color="#777777")
    
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('green')
```

```python
plt.figure(figsize=(12,10))
for i in range(10):
    plt.subplot(5,4,2*i+1)
    plot_image(i+50, predictions[i+50], label_test, data_test)
    plt.subplot(5,4,2*i+2)
    plot_value_array(i+50, predictions[i+50],  label_test)
plt.show()
```
